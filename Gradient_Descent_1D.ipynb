How ML Works:

2. compute error
3. Learn from mistakes & Modify parameters to minimize the errors
         How Through Calculas
         

```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
#define function
def fx(x):
  return x**2-2*x+3

#define derivative of the above function
def deriv(x):
  return 2*x-2
```

```python
#plot function and its derivative
x=np.linspace(-4,4,2000)

#plot
plt.plot(x,fx(x),x,deriv(x))
plt.grid()

plt.show()
```

# Implement Gradient descent

```python
locmin=np.random.choice(x,1)

#learning parameters
lr=.01
epochs=500
print(locmin)
#implementation
for i in range(epochs):
  grad=deriv(locmin)
  locmin=locmin-lr*grad

locmin
```

```python
#Plot
#plot
plt.plot(x,fx(x),x,deriv(x))
plt.plot(locmin,deriv(locmin),'ro')
plt.grid()
#plt.ylim((-.0001,.0008))
plt.show()
```
