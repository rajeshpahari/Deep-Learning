```python
#Import libraries
import torch
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':18})
```

```python
#input Variable
x=torch.linspace(-3,3,101)

#Create a function that returns activated output
def NNoutputx(actfun):
  #Get activation function type
  #This code replaces torch.relu with torch.<actfun>
  actfun=getattr(torch,actfun)
  return actfun(x)

```

```python
#activation functions
act_funcs=['relu','sigmoid','tanh']

fig=plt.figure(figsize=(10,8))

for actfun in act_funcs:
  plt.plot(x,NNoutputx(actfun),label=actfun,lw=3)

plt.xlabel('x')
plt.legend()
plt.ylim([-3,3])
plt.grid()
plt.show()
```

```python
#input Variable
x=torch.linspace(-3,3,101)

#Create a function that returns activated output
def nnoutputx(actfun):
  #Get activation function type
  #This code replaces torch.relu with torch.<actfun>
  actfun=getattr(torch.nn,actfun)  # See torch.nn here
  return actfun() # Not retrning actfun(x) but actfun()

```

```python
#activation functions
act_funcs=['ReLU6','Hardshrink','ReLU']

fig=plt.figure(figsize=(10,8))

for actfun in act_funcs:
  plt.plot(x,nnoutputx(actfun)(x),label=actfun,lw=3)# Applying x here

plt.xlabel('x')
plt.legend()
plt.ylim([-3,3])
plt.grid()
plt.show()
```
