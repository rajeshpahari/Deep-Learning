```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
x= np.linspace(-2*np.pi,2*np.pi,401)
fx=np.sin(x)*np.exp(-x**2*.05)
#derivative of fx
df=np.cos(x)*np.exp(-x**2*.05)+np.sin(x)*(-.1*x)*np.exp(-x**2*.05)


plt.plot(x,fx,x,df)
plt.legend(['f(x)','df'])
```

```python
# Lets define the function
def fx(x):
  return np.sin(x)*np.exp(-x**2*.05)

def deriv(x):
  return np.cos(x)*np.exp(-x**2*.05)+np.sin(x)*(-.1*x)*np.exp(-x**2*.05)

plt.plot(x,fx(x),x,deriv(x))
```

```python
# Gradient descent

#start point
locmin=np.random.choice(x,1)

#learning parameters
lr=.01
epochs=1000

#iteration
for i in range(epochs):
  grad=deriv(locmin)
  locmin=locmin-lr*grad


#plot the results
plt.plot(x,fx(x),x,deriv(x),'--')
plt.plot(locmin,deriv(locmin),'ro')
plt.plot(locmin,fx(locmin),'ro')

plt.grid()
plt.show()
```

So Gradient descent not always gives you global minima


Now we will do parametric experiment

```python
#Exp1: Change start position

startlocs=np.linspace(-5,5,1000)
finalres=np.zeros(len(startlocs))


#Loop for start points
for idx,locmin in enumerate(startlocs):

  #Run training

  for i in range(epochs):
    grad=deriv(locmin)
    locmin=locmin-lr*grad


  finalres[idx]=locmin

plt.plot(startlocs,finalres,'s-')
plt.show()

```

Global minima is between -4 to 1.5

```python
#Exp: Vary lr & Epochs
lrs=np.linspace(1e-10,1e-1,50)

finalres=np.zeros(len(lrs))


#Loop for start points
for idx,lr in enumerate(lrs):

  #Run training
  locmin=0
  for i in range(epochs):
    grad=deriv(locmin)
    locmin=locmin-lr*grad


  finalres[idx]=locmin

plt.plot(lrs,finalres,'s-')
plt.xlabel("LR")
plt.ylabel("Final Guess")
plt.show()
```
